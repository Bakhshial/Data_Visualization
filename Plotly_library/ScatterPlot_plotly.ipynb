```python
fig=px.scatter(
    df,
    x=df.Sales,
    y=df.Profit,
    color="Order Priority",
    title="Sales vs Profit index",
    symbol="Order Priority",
    marginal_x="box",
    marginal_y="box"
)
fig.show()
```

```python
fig=px.scatter(
    df,
    x=df.Sales,
    y=df.Profit,
    color="Order Priority",
    title="Sales vs Profit index",
    symbol="Order Priority",
    opacity=0.5,
    marginal_x="histogram",
    marginal_y="box"
)
fig.show()
```

```python
fig=px.scatter(
    df,
    x=df.Sales,
    y=df.Profit,
    color="Order Priority",
    title="Sales vs Profit index",
    symbol="Order Priority",
    facet_col="Order Priority",
    marginal_x="violin",
    marginal_y="violin"
)
fig.show()
```

```python
fig=px.scatter(
    df,
    x=df.Sales,
    y=df.Profit,
    color="Order Priority",
    title="Sales vs Profit index",
    symbol="Order Priority",
    marginal_x="box",
    marginal_y="rug"
)
fig.show()
```
